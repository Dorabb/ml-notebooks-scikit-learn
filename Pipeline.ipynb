# Pipeline: chaning pre-processors and estimators

```python
#Unified Transformers and estimator = a Pipeline object
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# create a pipeline object
pipe = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)
```

```python
# load the iris dataset and split it into train and test tests
X, y = load_iris(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# fit the whole pipeline
pipe.fit(X_train, y_train)
```

```python
#Pipeline(steps=[('standardscalar', StandardScaler()),('logisticregression', LogisticRegression())])
#we can now use it likeany other estimator
accuracy_score(pipe.predict(X_test), y_test)
```
